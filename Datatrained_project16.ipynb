```python
from sklearn.linear_model import LinearRegression
```

```python
Z = df[['Transportation expense','Age','Son','Body mass index']]
y = df['Absenteeism time in hours']
```

```python
lm = LinearRegression()
lm
```

```python
lm.fit(Z,y)
```

```python
Yhat=lm.predict(Z)
Yhat[0:5]   
```

```python
lm.fit(Z, y)
print('The R-square is: ', lm.score(Z, y))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
```

```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False,degree=6)), ('model',LinearRegression())]
```

```python
pipe=Pipeline(Input)
pipe
```

```python
pipe.fit(Z,y)
```

```python
ypipe=pipe.predict(Z)
ypipe[0:4]
```

```python
pipe.fit(Z, y)
# Find the R^2
print('The R-square is: ', pipe.score(Z, y))
```
